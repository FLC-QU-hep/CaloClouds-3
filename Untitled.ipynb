```python
fig, axarr = plt.subplots(1, 3, figsize=(15, 3))


for a, ax in enumerate(axarr):
    if cc_version == 3:
        i = a
    else:
        i = a+2
    cog = data[:, i]
    input_cog = input_events[:, i].detach().numpy()

    n, _, _ = ax.hist(cog, histtype='step', bins=50, label=f'model {np.mean(cog):.2}, {np.std(cog):.2}')
    n, _, _ = ax.hist(input_cog, histtype='step', bins=50, label=f'input {np.mean(input_cog):.2}, {np.std(input_cog):.2}')
    ax.semilogy()
    ax.legend()

weird_events = np.where((np.abs(cog) > 4))[0]
print(f"Of {len(data)} events {len(weird_events)} have CoG > 4, that is {len(weird_events)/len(data):%}")
```

```python
fig, axarr = plt.subplots(1, 1, figsize=(15, 5))

qs = [0, 0.1, 0.25, 0.5, 0.75, 0.9, 1]
model_quantiles = np.clip(np.quantile(data, qs, axis=0), -10, 10)
input_quantiles = np.clip(np.quantile(input_events, qs, axis=0), -10, 10)
xs = np.arange(len(model_quantiles[0]))
xs = np.tile(xs, len(qs))
c = np.repeat(qs, len(model_quantiles[0]))
    
points = axarr.scatter(xs, model_quantiles.flatten(), c=c, cmap='winter', label='model')

axarr.scatter(xs+0.5, input_quantiles.flatten(), c=c, cmap='spring', label='input')
plt.legend()
```
